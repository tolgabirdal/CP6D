```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
exp = 'experiments/posenet_results_nc_score_norm_debug/'

df_7scenes_trans = pd.read_csv(exp + '7Scenes_trans.csv')
df_7scenes_rot = pd.read_csv(exp + '7Scenes_rot.csv')
df_cam_trans = pd.read_csv(exp + 'CambridgeLandmarks_trans.csv')
df_cam_rot = pd.read_csv(exp + 'CambridgeLandmarks_rot.csv')
df_pho_trans = pd.read_csv(exp + 'PhotoTourism_trans.csv')
df_pho_rot = pd.read_csv(exp + 'PhotoTourism_rot.csv')
```

```python
dfs = [[df_7scenes_trans, df_7scenes_rot], [df_cam_trans, df_cam_rot], [df_pho_trans, df_pho_rot]]
x = np.linspace(0, 1, 50)
# Create a figure and two subplots with shared x and y axes
fig, axs = plt.subplots(2, 3, sharex=True, sharey=True, figsize=(25, 20))
font = 30
axs[0, 0].set_title('(1) 7Scenes', fontsize=font)
axs[0, 1].set_title('(2) CambridgeLandmarks', fontsize=font)
axs[0, 2].set_title('(3) PhotoTourism', fontsize=font)
axs[1, 0].set_xlabel('Percentage', fontsize=font)
axs[1, 1].set_xlabel('Percentage', fontsize=font)
axs[1, 2].set_xlabel('Percentage', fontsize=font)
axs[0, 0].set_ylabel('(a) Normalized Mean Trans Err', fontsize=font)
axs[1, 0].set_ylabel('(b) Normalized Mean Rot Err', fontsize=font)
for i in range(3):
    for j in range(2):
        for columns in dfs[i][j]:
            y = np.array(dfs[i][j][columns])
            y = (y - np.min(y)) / (np.max(y) - np.min(y))
            axs[j, i].plot(x, y, label=columns)
            axs[j, i].tick_params(axis='both', which='major', labelsize=font-2)
            axs[j, i].legend(fontsize=font-2)
fig.tight_layout()
plt.savefig('main_posenet_normalize_nc_score_debug.pdf')
```
