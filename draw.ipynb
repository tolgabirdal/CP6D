```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_7scenes_trans = pd.read_csv('7Scenes_trans.csv')
df_7scenes_rot = pd.read_csv('7Scenes_rot.csv')
df_cam_trans = pd.read_csv('CambridgeLandmarks_trans.csv')
df_cam_rot = pd.read_csv('CambridgeLandmarks_rot.csv')
df_pho_trans = pd.read_csv('PhotoTourism_trans.csv')
df_pho_rot = pd.read_csv('PhotoTourism_rot.csv')
```

```python
dfs = [[df_7scenes_trans, df_7scenes_rot], [df_cam_trans, df_cam_rot], [df_pho_trans, df_pho_rot]]
x = np.linspace(0, 1, 50)
# Create a figure and two subplots with shared x and y axes
fig, axs = plt.subplots(2, 3, sharex=True, sharey=True, figsize=(15, 10))
font = 16
axs[0, 0].set_title('7Scenes', fontsize=font)
axs[0, 1].set_title('CambridgeLandmarks', fontsize=font)
axs[0, 2].set_title('PhotoTourism', fontsize=font)
axs[1, 0].set_xlabel('Percentage', fontsize=font)
axs[1, 1].set_xlabel('Percentage', fontsize=font)
axs[1, 2].set_xlabel('Percentage', fontsize=font)
axs[0, 0].set_ylabel('Normalized Mean Trans Err', fontsize=font)
axs[1, 0].set_ylabel('Normalized Mean Rot Err', fontsize=font)
for i in range(3):
    for j in range(2):
        for columns in dfs[i][j]:
            y = np.array(dfs[i][j][columns])
            y = (y - np.min(y)) / (np.max(y) - np.min(y))
            axs[j, i].plot(x, y, label=columns)
            axs[j, i].tick_params(axis='both', which='major', labelsize=font-2)
            axs[j, i].legend(fontsize=font-1)
fig.tight_layout()
plt.savefig('main.pdf')
```
